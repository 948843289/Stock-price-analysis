```
print(df.index)##The date corresponding to the stock price
```

```
df['High-Low'] = df['High']-df['Low']#new row High-Low
```

```
df.head()
```

```
df.apply(np.mean)##Average each column of the data
```

```
change = df.Close.diff()#Add a column of change to store the value of the increase or 
#decrease between the stock price of the day and the closing price of the previous day,
#that is, the difference between the Close price of the current day and the Close price of the previous day
df['Change'] = change
df.head(5)
```

```
change.fillna(change.mean(),inplace=True)#For missing data, replace NaN with the mean of the rise and fall values.
```

```
df['pct_change'] = (df['Change'] /df['Close'].shift(1))#To calculate the fluctuation range, the idea of the pct_change() 
#algorithm is that the second item starts to subtract forward and then divides by the first item to calculate the 
#fluctuation sequence.
df['pct_change1'] = df.Close.pct_change()
df.head()
```

```
import matplotlib.pyplot as plt
df['Adj Close'].plot(legend=True, figsize=(10,4))##Stock price changes
plt.show()
```

```
df['pct_change'].plot(legend=True, figsize=(10,4))#Rise and fall
plt.show()
```

```
stock = df['Adj Close']
daily_return = (stock.diff()/stock.shift(periods = 1)).dropna()#Daily rate of return
daily_return.head()
```

```
import seaborn as sns
fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(15,6))
sns.distplot(daily_return)
ax.set_title("Alibaba")
```

```
company = ['BABA','GOOG','MSFT','AMZN','FB']#Multiple company stocks
top_tech_df = pdr.get_data_yahoo(company)['Adj Close']
top_tech_df.head()
```

```
top_tech_dr = top_tech_df.pct_change()#Rise and fall
top_tech_dr.head()
```

```
sns.jointplot('BABA','AMZN', top_tech_dr, kind='scatter')#Draw a scatter plot of the closing price difference between Alibaba and 
#Amazon, and look at the correlation.
```
